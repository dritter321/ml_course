```python
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()
import matplotlib.pyplot as plt
```

**Read data**

```python
train_df = pd.read_csv("https://raw.githubusercontent.com/dritter321/ml_course/main/data/titanic_train.csv", index_col="PassengerId")
```

```python
train_df.head(2)
```

```python
train_df.describe(include="all")
```

```python
train_df.info()
```

**Let's drop`Cabin`, and then â€“ all rows with missing values.**

```python
train_df = train_df.drop("Cabin", axis=1).dropna()
```

```python
train_df.shape
```

**1. Build a picture to visualize all scatter plots for each pair of features `Age`, `Fare`, `SibSp`, `Parch` and `Survived`. ( `scatter_matrix ` from Pandas or `pairplot` from Seaborn)**

```python
sns.pairplot(train_df)
```

**2. How does ticket price (`Fare`) depend on `Pclass`? Build a boxplot.**

```python
# Create a boxplot using Seaborn
plt.figure(figsize=(20, 4))  # Set the figure size
sns.boxplot(x='Fare', y='Pclass', data=train_df, palette='Set3', orient="h", hue='Pclass', legend=False)

# Add title and labels
plt.title('Relationship between Fare and Pclass')
plt.xlabel('Fare')
plt.ylabel('Pclass')

# Show the plot
plt.show()
```

**3. Let's build the same plot but restricting values of `Fare` to be less than 95% quantile of the initial vector (to drop outliers that make the plot less clear).**

```python
# Calculate the 95th percentile of Fare
fare_95th_percentile = train_df['Fare'].quantile(0.95)

# Filter the data to include only Fare values less than the 95th percentile
filtered_train_df = train_df[train_df['Fare'] < fare_95th_percentile]

# Create a boxplot using Seaborn
plt.figure(figsize=(20, 4))  # Set the figure size
sns.boxplot(x='Fare', y='Pclass', data=filtered_train_df, palette='Set3', orient="h", hue='Pclass', legend=False)

# Add title and labels
plt.title('Relationship between Fare and Pclass (Fare < 95th Percentile)')
plt.xlabel('Fare')
plt.ylabel('Pclass')

# Show the plot
plt.show()
```

**4. How is the percentage of surviving passengers dependent on passengers' gender? Depict it with `Seaborn.countplot` using the `hue` argument.**

```python
# Calculate the percentage of surviving passengers for each gender
survival_percentage_by_gender = train_df.groupby(['Sex', 'Survived']).size().unstack() * 100 / train_df.groupby('Sex').size()

# Plot the data using Seaborn countplot
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=train_df)

# Add title and labels
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Show the plot
plt.legend(title='Survived', labels=['No', 'Yes'])  # Adding legend
plt.show()
```

**5. How does the distribution of ticket prices differ for those who survived and those who didn't. Depict it with `Seaborn.boxplot`**

```python
# Filter the data for passengers who survived and those who didn't
survived_data = train_df[train_df['Survived'] == 1]
not_survived_data = train_df[train_df['Survived'] == 0]

# Create a boxplot using Seaborn
plt.figure(figsize=(20, 6))
sns.boxplot(x='Fare', y='Survived', data=train_df, orient="h")

# Add title and labels
plt.title('Distribution of Ticket Prices for Survived and Not Survived Passengers')
plt.xlabel('Fare')
plt.ylabel('Survived')

# Show the plot
plt.show()
```

**6. How does survival depend on passengers' age?  Verify (graphically) an assumption that youngsters (< 30 y.o.) survived more frequently than old people (> 55 y.o.).**

```python
# Define age categories: Youngsters (< 30), Middle-aged (30-55), Old (> 55)
train_df['Age_Category'] = pd.cut(train_df['Age'], bins=[0, 30, 55, train_df['Age'].max()], labels=['Youngsters', 'Middle-aged', 'Old'])

# Create a countplot to visualize survival based on age categories
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_Category', hue='Survived', data=train_df, palette='Set2')

# Add title and labels
plt.title('Survival Count by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Count')

# Show the plot
plt.legend(title='Survived', labels=['No', 'Yes'])  # Adding legend
plt.show()
```
